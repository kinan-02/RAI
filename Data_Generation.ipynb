```python
import numpy as np

def calculate_total_surgeries(row):
    return sum(workplace['Number of Surgeries Performed'] for workplace in row['Workplaces'])

df['Total_Number of Surgeries Performed'] = df.apply(calculate_total_surgeries, axis=1)

def calculate_avg(row, title):
    return sum(workplace[title] for workplace in row['Workplaces'])/len(row['Workplaces'])

def complex_acceptance_criteria(row):
    weight_experience = 0.2
    weight_surgeries = 0.15
    weight_malpractice = -0.3
    weight_language = 0.1
    weight_recommendations = 0.1
    weight_patient_feedback = 0.1
    weight_employer_recommendations = 0.1
    weight_insurance_history = 0.1
    weight_exam_results = 0.2

    experience_score = min(row['Total Years of Experience'] / 40, 1) * weight_experience
    surgery_score = min(row['Total_Number of Surgeries Performed'] / 1000, 1) * weight_surgeries
    malpractice_score = (1 - min(row['Medical Malpractice Claims'] / 10, 1)) * weight_malpractice
    language_score = (['Beginner', 'Intermediate', 'Advanced', 'Fluent'].index(row['Proficiency in Hebrew']) / 3 +
                      ['Beginner', 'Intermediate', 'Advanced', 'Fluent'].index(row['Proficiency in English']) / 3) / 2 * weight_language

    recommendation_label = calculate_avg(row, 'Recommendation Letters Label')
    recommendation_score = (recommendation_label / 5) * weight_recommendations

    patient_feedback_label = calculate_avg(row, 'Patient Feedback Label')
    patient_feedback_score = (patient_feedback_label / 5) * weight_patient_feedback

    employer_recommendations_label = calculate_avg(row, 'Recommendations from Former Employers Label')
    employer_recommendations_score = (employer_recommendations_label / 5) * weight_employer_recommendations

    insurance_history_label = row['Insurance History Label']
    insurance_history_score = (insurance_history_label / 5) * weight_insurance_history

    exam_score = (row['Israeli Medical Licensing Exam Results'] / 100) * weight_exam_results

    total_score = (experience_score + surgery_score + malpractice_score + language_score +
                   recommendation_score + patient_feedback_score + employer_recommendations_score +
                   insurance_history_score + exam_score)

    acceptance_probability = 1 / (1 + np.exp(-total_score))  # Sigmoid function to map to probability

    return 1 if acceptance_probability > 0.5 else 0

df['Accepted'] = df.apply(complex_acceptance_criteria, axis=1)

# Display the updated DataFrame
# import ace_tools as tools; tools.display_dataframe_to_user(name="Surgeons Acceptance Data", dataframe=df)
```
